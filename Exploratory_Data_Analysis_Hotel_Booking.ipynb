## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# drop the duplicate data
dataset.drop_duplicates(inplace=True)
````

````python
# check the data duplicate value present or not
dataset.duplicated().sum()
````

````python
# handle null value that showing already
missing_columns
````

````python
# Fill null values in children column with 0
dataset['children'].fillna(0,inplace=True)

# Fill null values in country column with unknown
dataset['country'].fillna('unknown',inplace=True)

#Fill null values in agent column with 0
dataset['agent'].fillna(0,inplace=True)

# drop the company column
dataset.drop(columns='company', inplace = True,axis = 0)
````

````python
# check the shape of dataset
dataset.shape
````

````python
# further check the null / missing value
dataset.isnull().sum()
````

````python
# manipulate the data
# showing the unique month name in arrival date month column
dataset['arrival_date_month'].unique()
````

````python
#Now we will Create a new column season
dataset['season'] = dataset['arrival_date_month']
dataset['season'] = dataset['season'].replace(['October', 'November', 'December','January', 'February'],'Winter')
dataset['season'] = dataset['season'].replace(['March', 'April', 'May', 'June','July', 'August', 'September'],'Summer')
````

````python
#creating a new column date
dataset['date'] = dataset['arrival_date_day_of_month'].astype(str) + '/' + dataset['arrival_date_month'].astype(str) + '/' + dataset['arrival_date_year'].astype(str)
dataset['date'] = pd.to_datetime(dataset['date'], format='%d/%B/%Y')
````

````python
dataset
````

````python
#Converting data type of agent column from float to int
dataset['agent'] = dataset['agent'].astype(int)
````
